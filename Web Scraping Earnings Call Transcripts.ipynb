```python
import requests
from bs4 import BeautifulSoup
import re
import nltk
from nltk import word_tokenize
import csv
from collections import Counter
import string
import pandas as pd
from openpyxl import Workbook
```

```python
def get_text(link, ad):
    requests.packages.urllib3.disable_warnings()
    url = link
    response = requests.get(url, verify=False)
    if not response.status_code == 200:
        None
    results = BeautifulSoup(response.content, 'lxml')
    text = results.find('div',class_='main-container').get_text().replace('\n','').split('Prepared Remarks:')[1].split('[Operator signoff]')[0].replace('Operator','').replace('\xa0', '').strip()
    text = re.sub(re.escape(ad), '', text)
    return text
```

```python
dlr_link = 'https://www.fool.com/earnings/call-transcripts/2023/04/28/digital-realty-trust-dlr-q1-2023-earnings-call-tra/'
dlr_ad = "10 stocks we like better thanDigital Realty TrustWhen our analyst team hasa stock tip, it can pay to listen. After all, the newsletter theyhave run for over a decade, Motley Fool Stock Advisor, has tripled the market.*Theyjust revealed what they believe are thetenbest stocksfor investors to buy right now... and Digital Realty Trustwasn't one of them! That's right -- theythink these 10stocks are even better buys.See the 10 stocks*Stock Advisor returns as of April 24, 2023Before I turn the call over to Andy, let me offer a few key takeaways from our first quarter."
dlr_text = get_text(dlr_link, dlr_ad)
dlr_text
```

```python
pld_link = 'https://www.fool.com/earnings/call-transcripts/2023/01/18/prologis-pld-q4-2022-earnings-call-transcript/'
pld_ad = "10 stocks we like better thanPrologisWhen our award-winning analyst team hasa stock tip, it can pay to listen. After all, the newsletter theyhave run for over a decade, Motley Fool Stock Advisor, has tripled the market.*Theyjust revealed what they believe are thetenbest stocksfor investors to buy right now... and Prologiswasn't one of them! That's right -- theythink these 10stocks are even better buys.See the 10 stocks*Stock Advisor returns as of January 9, 2023"
pld_text = get_text(pld_link, pld_ad)
pld_text
```

```python
def word_list(text):
    word_list = []
    filler_words = ['oh', 'like', 'well', 'just', 'basically', 'actually', 'so', 'right', 'okay', 'alright', 'um', 'yeah', 'pretty']
    text = text.translate(str.maketrans('', '', string.punctuation))
    for word in word_tokenize(text.lower()):
        if word not in filler_words:
            word_list.append(word)
    return word_list
```

```python
dlr_word_list = word_list(dlr_text)
dlr_word_list
```

```python
pld_word_list = word_list(pld_text)
pld_word_list
```

```python
def get_pos_neg_words():
    def get_words(url):
        import requests
        words = requests.get(url, verify=False).content.decode('latin-1')
        word_list = words.split('\n')
        index = 0
        while index < len(word_list):
            word = word_list[index]
            if ';' in word or not word:
                word_list.pop(index)
            else:
                index+=1
        return word_list
    p_url = 'http://ptrckprry.com/course/ssd/data/positive-words.txt'
    n_url = 'http://ptrckprry.com/course/ssd/data/negative-words.txt'
    positive_words = get_words(p_url)
    negative_words = get_words(n_url)
    return positive_words,negative_words
```

```python
def sentiment_analysis(word_list):
    pos_words, neg_words = get_pos_neg_words()
    pos = set()
    neg = set()
    for word in word_list:
        if word in pos_words:
            if word not in pos:
                pos.add(word)
        if word in neg_words:
            if word not in neg:
                neg.add(word)
    return pos, len(pos), round(len(pos)/len(word_list)*100,2), neg, len(neg), round(len(neg)/len(word_list)*100,2)
```

```python
def word_count(word_list):
    pos_words, neg_words = get_pos_neg_words()
    pos = []
    neg = []
    for word in word_list:
        if word in pos_words:
            pos.append(word)
        if word in neg_words:
            neg.append(word)
    pos_top = sorted(Counter(pos).items(), key=lambda x: x[1], reverse=True)
    neg_top = sorted(Counter(neg).items(), key=lambda x: x[1], reverse=True)
    return pos_top, neg_top
```

```python
def save_file(file_name, word_list):
    with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
        wb = writer.book
        pd.DataFrame(sentiment_analysis(word_list)[0]).to_excel(writer, sheet_name='Sheet1', index=False)
        worksheet1 = wb['Sheet1']
        worksheet1.title = 'List of positive words'
        worksheet1['A1'] = 'Word'
        worksheet1['C1'] = 'Number of positive words'
        worksheet1['C2'] = sentiment_analysis(word_list)[1]
        worksheet1['C4'] = 'Ratio of positive words'
        worksheet1['C5'] = sentiment_analysis(word_list)[2]
        pd.DataFrame(sentiment_analysis(word_list)[3]).to_excel(writer, sheet_name='Sheet2', index=False)
        worksheet2 = wb['Sheet2']
        worksheet2.title = 'List of negative words'
        worksheet2['A1'] = 'Word'
        worksheet2['C1'] = 'Number of negative words'
        worksheet2['C2'] = sentiment_analysis(word_list)[4]
        worksheet2['C4'] = 'Ratio of negative words'
        worksheet2['C5'] = sentiment_analysis(word_list)[5]
        pd.DataFrame(word_count(word_list)[0]).to_excel(writer, sheet_name='Sheet3', index=False)
        worksheet3 = wb['Sheet3']
        worksheet3.title = 'List of positive words and count'
        worksheet3['A1'] = 'Word'
        worksheet3['B1'] = 'Count'
        pd.DataFrame(word_count(word_list)[1]).to_excel(writer, sheet_name='Sheet4', index=False)
        worksheet4 = wb['Sheet4']
        worksheet4.title = 'List of negative words and count'
        worksheet4['A1'] = 'Word'
        worksheet4['B1'] = 'Count'
        pd.DataFrame(word_count(word_list)[0][:10]).to_excel(writer, sheet_name='Sheet5', index=False)
        worksheet5 = wb['Sheet5']
        worksheet5.title = 'Top 10 positive words and count'
        worksheet5['A1'] = 'Word'
        worksheet5['B1'] = 'Count'
        pd.DataFrame(word_count(word_list)[1][:10]).to_excel(writer, sheet_name='Sheet6', index=False)
        worksheet6 = wb['Sheet6']
        worksheet6.title = 'Top 10 negative words and count'
        worksheet6['A1'] = 'Word'
        worksheet6['B1'] = 'Count'
        writer.save()
    return f"Output saved to '{file_name}' successfully."
```
